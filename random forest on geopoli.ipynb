```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
df=pd.read_excel("geopoli
                 .xlsx")
df
```

```python
# Replace missing values with 0
df['Total Deaths'].fillna(0, inplace=True)
df['No Injured'].fillna(0, inplace=True)
```

```python
y=df.Previous_Disaster
encoder = LabelEncoder()
encoded_values=encoder.fit_transform(y)
df['Previous_Disaster'] = encoded_values
original_labels_dis = encoder.classes_
custom_mapping = {label: value for value, label in enumerate(original_labels_dis)}
print("Custom Mapping:")
print(custom_mapping)
```

```python
y=df.Country
encoder = LabelEncoder()
encoded_values=encoder.fit_transform(y)
df['Country'] = encoded_values
original_labels_count = encoder.classes_
custom_mapping = {label: value for value, label in enumerate(original_labels_count)}
print("Custom Mapping:")
print(custom_mapping)
```

```python
y=df.Disaster_Group
encoder = LabelEncoder()
encoded_values=encoder.fit_transform(y)
df['Disaster_Group'] = encoded_values
original_labels_G = encoder.classes_
custom_mapping = {label: value for value, label in enumerate(original_labels_G)}
print("Custom Mapping:")
print(custom_mapping)
```

```python
y=df.Origin
encoder = LabelEncoder()
encoded_values=encoder.fit_transform(y)
df['Origin'] = encoded_values
original_labels = encoder.classes_
custom_mapping = {label: value for value, label in enumerate(original_labels)}
print("Custom Mapping:")
print(custom_mapping)
```

```python
y=df.Region
encoder = LabelEncoder()
encoded_values=encoder.fit_transform(y)
df['Region'] = encoded_values
original_labels = encoder.classes_
custom_mapping = {label: value for value, label in enumerate(original_labels)}
print("Custom Mapping:")
print(custom_mapping)
```

```python
y=df.Location
encoder = LabelEncoder()
encoded_values=encoder.fit_transform(y)
df['Location'] = encoded_values
original_labels = encoder.classes_
custom_mapping = {label: value for value, label in enumerate(original_labels)}
print("Custom Mapping:")
print(custom_mapping)
```

```python
df
```

```python
X=df[['Year','Disaster_Group',"Country","Region","Location","Origin","Year_previous_disaster","End_Year_previous_disaster"]]
y=df[["Previous_Disaster",'Total Deaths','No Injured']]
```

```python
reg = RandomForestClassifier(min_samples_split=10, n_estimators=25, n_jobs=-1, random_state=0, max_depth=15, verbose=2)
reg = reg.fit(X, y)    
y_pred = reg.predict(X)
```

```python
y_pred
y_pred.shape
```

```python
df1=pd.read_excel("geopoli_org.xlsx")
results_df = df1.copy()  # Create a copy of the original DataFrame
x=y_pred[:, 0]
x = x.astype(int)
decoded_label_dis = original_labels_dis[x]
results_df['Future_Disaster'] = decoded_label_dis
results_df['total deaths predicted'] = y_pred[:, 1]
results_df['total injured predicted'] = y_pred[:, 2]
results_df.to_csv('predictions.csv', index=False)
```
